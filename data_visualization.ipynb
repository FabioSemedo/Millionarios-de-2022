```python
import pandas as pd

df = pd.read_csv('Billionaires.csv') #defines df as the dataset in study

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_ignore = ['full_name', 'first_name', 'last_name', 'birth_date', 'source', 'position']

columns_as_list = ['country_of_residence', 'city_of_residence', 'citizenship']

columns_pie_chart = ['continent', 'gender']

for column in df.columns:
    if column in columns_to_ignore:
        print(f"{column} ignored")
        continue
    
    if column in columns_as_list: #creates the lists
        print(f"{column} value count:")
        print(df[column].value_counts())
        print("-" * 50)
        continue
    
    if column in ['residence_state', 'residence_region']: #ignores the following value in the us subdivision
        filtered_data = df[df[column] != 'No subdivisions info']
        plt.figure(figsize=(25, 6))
        sns.countplot(x=filtered_data[column])
        plt.title(f"Column (Filtered): {column}")
        plt.xticks(rotation=45)
        plt.show()
        continue
    
    if column in columns_pie_chart:
        plt.figure(figsize=(10, 10))
        
        data_counts = df[column].value_counts()
        wedges, texts, autotexts = plt.pie(
            data_counts,
            labels=data_counts.index,
            autopct='%1.1f%%',
            startangle=90,
            textprops={'fontsize': 10},
            rotatelabels=True
        )
        for text in texts:
            text.set_rotation_mode('anchor')
            text.set_ha('center')
            text.set_va('center')
        
        plt.title(f"Column: {column}")
        plt.show()
    
    if df[column].dtype in ['int64', 'float64']: #graphs for numerical values - boxplots
            plt.figure(figsize=(25, 6))
            sns.boxplot(x=df[column])
            plt.title(f"Column: {column}")
            plt.show()
    
    else: #graphs for categorical values - countplots
            plt.figure(figsize=(25, 6))
            sns.countplot(x=df[column])
            plt.title(f"Column: {column}")
            plt.xticks(rotation=45)
            plt.show()
```
